```python
import time
import os
```

```python
def calculate_current(actual_voltage):
  primary_enroll = 1
  secundary_enroll = 160
  external_current = actual_voltage *5
  internal_current = actual_voltage *3
  time.strftime('%H:%M:%S')
  os.environ['TZ'] = 'America/Sao Paulo'
  time.tzset()
  format_time = time.strftime('%H:%M:%S')

  secundary_cur_external = (primary_enroll/secundary_enroll)*external_current
  secundary_cur_external = (primary_enroll/secundary_enroll)*internal_current

  return {
      'primary_current':actual_voltage,
      'time': format_time,
      'external_current':secundary_cur_external,
      'internal_current':secundary_cur_external,
  }
```

```python
sample = 1_0000
curr = []
for i in range(1,sample,10):
  actual_voltage = i
  curr.append(calculate_current(actual_voltage))
```
